```python
import labels as labels
```

```python
category = []
for label in labels.labels:
    x = {'id':label.id, 'name':label.name, 'catgeory': label.category, 
         'trainId':label.trainId}
    category.append(x)
```

```python
category
```

```python
def toSemanticImage(im):
        semanticImg = np.zeros(im.size)
        shape = im.shape
        im = im.flatten()
        for index in range(len(im)):
            semanticImg[index] = category[im[index]]['trainId']
        semanticImg = semanticImg.reshape(shape)
        return semanticImg
```

```python
from PIL import Image
count = 0
for train_mask in train_masks:
    file_path = data_path+'masktrain2014/'+train_mask[42:45]+train_mask[69:79]+'.jpg'
    im = cv.imread(data_path+train_mask, 0)
    mask = np.asarray(toSemanticImage(im), dtype ='uint8')
    mask = Image.fromarray(mask*255, mode = 'L')
    mask.save(file_path)
    count = count+1
print(count)
```

```python
count = 0
for val_mask in val_masks:
    file_path = data_path+'maskval2014/'+val_mask[42:45]+val_mask[69:79]+'.jpg'
    im = cv.imread(data_path+val_mask, 0)
    mask = np.asarray(toSemanticImage(im), dtype ='uint8')
    mask = Image.fromarray(mask*255, mode = 'L')
    mask.save(file_path)
    count = count+1
print(count)
```
